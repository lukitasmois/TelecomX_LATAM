# **ðŸ“Œ ExtraciÃ³n**

````python
import pandas as pd
````

````python
df = pd.read_json('https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json')
````

````python
df.head()
````

````python
df_norm = df[['customerID','Churn']]
````

````python
df_customers = pd.json_normalize(df['customer'])
df_norm[df_customers.columns] = df_customers
df_norm
````

````python
df_phones = pd.json_normalize(df['phone'])
df_norm[df_phones.columns] = df_phones
df_norm
````

````python
df_internets =  pd.json_normalize(df['internet'])
df_norm[df_internets.columns] = df_internets

df_accounts =  pd.json_normalize(df['account'])
df_norm[df_accounts.columns] = df_accounts

df_norm.head()
````

# **ðŸ”§ TransformaciÃ³n**

````python
import numpy as np
````

````python
df_norm[df_norm['Charges.Total'] == ' ']
````

````python
#Obtengo los indices de los registros vacios
indices_charges_0 = df_norm[df_norm['Charges.Total'] == ' '].index
````

````python
#Cambio el valor de los registros vacios por '0'
df_norm.loc[indices_charges_0, 'Charges.Total'] = 0
````

````python
#Verifico si siguen exisitiendo registros vacios
df_norm[df_norm['Charges.Total'] == ' ']
````

````python
#Cambio el tipo de dato de la columna 'Charges.Total'

df_norm['Charges.Total'] = df_norm['Charges.Total'].astype(np.float64)
````

````python
#Verifico el cambio el tipo de dato de la columna 'Charges.Total'
df_norm.info()
````

## Columna de cuentas diarias

````python
df_norm['Charges.Daily'] = df_norm['Charges.Monthly'] / 30
````

````python
df_norm.head()
````

## EstandarizaciÃ³n y transformaciÃ³n de datos

````python
# Creo una lista con las columnas a estandarizar
columns_to_standardize = ['Churn', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
'PaperlessBilling']
````

````python
# Cambio los valores 'Yes' y 'No' por valores booleanos
df_norm[columns_to_standardize] = df_norm[columns_to_standardize].replace('Yes',True).replace('No',False)
````

````python
df_norm.head()
````

## DistribuciÃ³n de evasiÃ³n

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
active_clients = df_norm[df_norm['Churn'] == True]['Churn'].count()
inactive_clients = df_norm[df_norm['Churn'] == False]['Churn'].count()

df_clients_status = pd.DataFrame({'Estado': ['Activo','Inactivo'], 'Valor': [active_clients, inactive_clients]})
````

````python
df_clients_status
````

````python
fig, ax = plt.subplots(figsize=(10, 6))
ax = sns.barplot(df_clients_status, x='Estado', y='Valor',hue='Valor', palette='tab10')

ax.set_title('Cantidad de empleados que dejaron y continuan en la empresa\n', fontsize=20)
ax.set_ylabel(' ')
ax.set_xlabel(' ')
ax.tick_params(labelsize=14)

ax.legend(title='Valores', loc='upper right', bbox_to_anchor=(1.2, 1), fontsize=14, title_fontsize=14)

plt.show()
````

````python
fig, ax = plt.subplots(figsize=(10,6))

plt.pie(df_clients_status['Valor'], labels=df_clients_status['Estado'], autopct='%0.f %%')
ax.set_title('Grafico de torta de los empleados que\n dejaron y continuan en la empresa con respecto al total')
plt.show()
````

## Recuento de evasiÃ³n por variables categÃ³ricas

````python
df_norm.head(12)
````

````python
columns_categorics =['gender', 'InternetService', 'Contract', 'PaymentMethod']
````

````python
df_categorics_churn = df_norm.loc[df_norm['Churn'] == True, columns_categorics]
df_categorics_churn.head()
````

````python
fig, axs = plt.subplots(2,2,figsize=(12,9))
fig.subplots_adjust(hspace=0.3, wspace=0.2)
axs = axs.flatten()
fig.suptitle('Recuento de evasiÃ³n por variables categÃ³ricas', fontsize=20)

for i, col in enumerate(columns_categorics):

  axs[i].pie(df_categorics_churn[col].value_counts(), labels=df_categorics_churn[col].unique(), autopct='%0.f %%', textprops={'fontsize':13})
  axs[i].set_title(f'Porcentaje de evacion segun la variable\n "{columns_categorics[i]}" de los/las clientes/as.', fontsize=16)
plt.show()
````

## Conteo de evasiÃ³n por variables numÃ©ricas

````python
numerical_columns = ['tenure', 'Charges.Monthly','Charges.Total','Charges.Daily']
````

````python
fig, axs = plt.subplots(2,2, figsize=(14,6))
fig.subplots_adjust(hspace=0.5, wspace=0.2)

fig.suptitle('Conteo de evasiÃ³n segun las variables numericas', fontsize=18)

axs = axs.flatten()

for i, col in enumerate(numerical_columns):
#   axs[i].boxplot(df_norm[df_norm['Churn'] == True][col])
  ax = sns.boxplot(df_norm, x='Churn', y=col, ax=axs[i], palette='tab10')
  axs[i].set_xlabel(' ')
  axs[i].tick_params(labelsize=12)
  axs[i].set_ylabel(col , fontsize=12)
  ax.grid()

plt.show()
````

## Introduccion

El objetivo del analisis de los datos de la tienda TelecomX es identificar la causa principal de evasion por parte de los clientes utilizando diferentes tipos de herramientas y graficos para visualizar el comportamiento segun las caracteristicas de los diferentes servicios contratados.

## Limpieza y Tratamiento de Datos

### Importacion de datos

Mediante la biblioteca "Pandas" se realizo la importacion del set de datos desde el link proporcionado por TelecomX.

### Tratamiento de datos

Una vez importados, se realizo una normalizacion de los datos para poder trabajar con los mismos de la forma correcta. Se creo un "DataFrame" para obtener un formato tabular, esto no solo permite realizar el analisis de forma mas eficiente, sino que tambien de forma mas ordenada.

### Analisis exploratorio de datos

Primero utilizamos ```df_norm.describe()``` para observar el comportamiento general de los clientes registrados, esten o no esten activos. Aqui podemos observar la tendencia de la eleccion de planes a largo plazo.

Luego utilizando un grafico de barras pudimos comparar la cantidad de clientes que estan activos (clientes que tienen un plan contratado) con los clientes inactivos (clientes que decidieron no continuar con el contrato del plan), gracias al cual pudimos observar que predominan los clientes inactivos:

````python
fig, ax = plt.subplots(figsize=(10, 6))
ax = sns.barplot(df_clients_status, x='Estado', y='Valor',hue='Valor', palette='tab10')

ax.set_title('Cantidad de empleados que dejaron y continuan en la empresa\n', fontsize=20)
ax.set_ylabel(' ')
ax.set_xlabel(' ')
ax.tick_params(labelsize=14)

ax.legend(title='Valores', loc='upper right', bbox_to_anchor=(1.2, 1), fontsize=14, title_fontsize=14)

plt.show()
````

Continuamos realizando un analisis sobre las variables categoricas sobre los clientes inactivos, observando el genero, tipo de servicio contratado, tipo de contrato y metodo de pago a eleccion. Gracias a estos graficos identificamos que la mayoria de clientes que decidieron dar de baja el servicio, contrataron el servicio de fibra optica, abonaban de forma electronica y elegian los contratos "mes a mes", haciendo enfasis en esta ultima categoria que se relaciona con el siguiente analisis.

````python
fig, axs = plt.subplots(2,2,figsize=(12,9))
fig.subplots_adjust(hspace=0.3, wspace=0.2)
axs = axs.flatten()
fig.suptitle('Recuento de evasiÃ³n por variables categÃ³ricas', fontsize=20)

for i, col in enumerate(columns_categorics):

  axs[i].pie(df_categorics_churn[col].value_counts(), labels=df_categorics_churn[col].unique(), autopct='%0.f %%', textprops={'fontsize':13})
  axs[i].set_title(f'Porcentaje de evacion segun la variable\n "{columns_categorics[i]}" de los/las clientes/as.', fontsize=16)
plt.show()
````

Luego de realizar el conteo de evasion segun las variables numericas, podemos observar como los clientes que mÃ¡s abandonaron suelen tener poco tiempo de contrato, lo que podrÃ­a indicar que los primeros meses del servicio son crÃ­ticos para la retenciÃ³n. Esto esta respaldado con el analisis anterior que visualiza que los clientes que mas abandonaron son los que eligieron contratos "mes a mes".

````python
fig, axs = plt.subplots(2,2, figsize=(14,6))
fig.subplots_adjust(hspace=0.5, wspace=0.2)

fig.suptitle('Conteo de evasiÃ³n segun las variables numericas', fontsize=18)

axs = axs.flatten()

for i, col in enumerate(numerical_columns):
#   axs[i].boxplot(df_norm[df_norm['Churn'] == True][col])
  ax = sns.boxplot(df_norm, x='Churn', y=col, ax=axs[i], palette='tab10')
  axs[i].set_xlabel(' ')
  axs[i].tick_params(labelsize=12)
  axs[i].set_ylabel(col , fontsize=12)
  ax.grid()

plt.show()
````
