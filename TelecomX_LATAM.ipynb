# **ðŸ“Œ ExtraciÃ³n**

```python
import pandas as pd
```

```python
df = pd.read_json('https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json')
```

```python
df.head()
```

```python
df_norm = df[['customerID','Churn']]
```

```python
df_customers = pd.json_normalize(df['customer'])
df_norm[df_customers.columns] = df_customers
df_norm
```

```python
df_phones = pd.json_normalize(df['phone'])
df_norm[df_phones.columns] = df_phones
df_norm
```

```python
df_internets =  pd.json_normalize(df['internet'])
df_norm[df_internets.columns] = df_internets

df_accounts =  pd.json_normalize(df['account'])
df_norm[df_accounts.columns] = df_accounts

df_norm.head()
```

```python
#Obtengo los indices de los registros vacios
indices_charges_0 = df_norm[df_norm['Charges.Total'] == ' '].index
```

```python
#Cambio el valor de los registros vacios por '0'
df_norm.loc[indices_charges_0, 'Charges.Total'] = 0
```

```python
#Verifico si siguen exisitiendo registros vacios
df_norm[df_norm['Charges.Total'] == ' ']
```

```python
#Cambio el tipo de dato de la columna 'Charges.Total'

df_norm['Charges.Total'] = df_norm['Charges.Total'].astype(np.float64)
```

```python
#Verifico el cambio el tipo de dato de la columna 'Charges.Total'
df_norm.info()
```

## Columna de cuentas diarias

```python
df_norm['Charges.Daily'] = df_norm['Charges.Monthly'] / 30
```

```python
df_norm.head()
```

## EstandarizaciÃ³n y transformaciÃ³n de datos

```python
# Creo una lista con las columnas a estandarizar
columns_to_standardize = ['Churn', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
'PaperlessBilling']
```

```python
# Cambio los valores 'Yes' y 'No' por valores booleanos
df_norm[columns_to_standardize] = df_norm[columns_to_standardize].replace('Yes',True).replace('No',False)
```

## DistribuciÃ³n de evasiÃ³n

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
active_clients = df_norm[df_norm['Churn'] == True]['Churn'].count()
inactive_clients = df_norm[df_norm['Churn'] == False]['Churn'].count()

df_clients_status = pd.DataFrame({'Estado': ['Activo','Inactivo'], 'Valor': [active_clients, inactive_clients]})
```

```python
df_clients_status
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax = sns.barplot(df_clients_status, x='Estado', y='Valor',hue='Valor', palette='tab10')

ax.set_title('Cantidad de empleados que dejaron y continuan en la empresa\n', fontsize=20)
ax.set_ylabel(' ')
ax.set_xlabel(' ')
ax.tick_params(labelsize=14)

ax.legend(title='Valores', loc='upper right', bbox_to_anchor=(1.2, 1), fontsize=14, title_fontsize=14)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))

plt.pie(df_clients_status['Valor'], labels=df_clients_status['Estado'], autopct='%0.f %%')
ax.set_title('Grafico de torta de los empleados que\n dejaron y continuan en la empresa con respecto al total')
plt.show()
```
